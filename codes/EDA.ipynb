```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = '../data/'

train_origin = pd.read_csv(data_path+'train_data.csv')
test_origin = pd.read_csv(data_path+'test_data.csv')
topic_dict_origin = pd.read_csv(data_path+'topic_dict.csv')
sample_submission_origin = pd.read_csv(data_path+'sample_submission.csv')
```

```python
train = train_origin.copy()
test = test_origin.copy()
topic_dict = topic_dict_origin.copy()
sample_submission = sample_submission_origin.copy()
```

```python
sns.countplot(data=train, x='topic_idx')
plt.show()
```

```python
length=train['title'].astype(str).apply(len)
plt.hist(length, bins=50, alpha=0.5, color='r', label='word')
plt.title('histogram of length of title')
plt.figure(figsize=(12, 5))
plt.boxplot(length, labels=['counts'], showmeans=True)
plt.show()
print('title 길이 최댓값: {}'.format(np.max(length)))
print('title 길이 최솟값: {}'.format(np.min(length)))
print('title 길이 평균값: {}'.format(np.mean(length)))
print('title 길이 중간값: {}'.format(np.median(length)))
```
